```python
path = '../data/usagov_bitly_data2012-03-16-1331923249.txt'
open(path).readline()


```

```python

import json
records = [json.loads(line) for line in open(path)] #list comprehension
records
```

```python
time_zones = [rec['tz'] for rec in records]

'''Oops! Turns out that not all of the records have a time zone field. 
This is easy to handle as we can add the check if 'tz' in rec at the end of the list comprehension: '''
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
```
